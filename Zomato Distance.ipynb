### Fetch the cuisine_id of of cuisine ‘BBQ’ using Zomato API.
### Print the cuisine_id

```python
import requests
header = {'user-key': '61b413ded5143f11855a455e7cf44568'}
response = requests.get('https://developers.zomato.com/api/v2.1/cuisines', headers = header, params = {'lat': 28.297850, 'lon': 77.140670})
data = response.json()
data
```

```python
cuisines = data['cuisines']
cuisines
```

```python
for i in range(0, len(cuisines)):
    if (cuisines[i]['cuisine']['cuisine_name'] == 'BBQ'):
        print(cuisines[i]['cuisine']['cuisine_id'])
        break
```

### Fetch the list of the top 10 restaurants serving cuisine ‘BBQ’ which are nearest to Coding Ninjas.
### Print the restaurant name, user rating, restaurant id and locality in which restaurant is present.

```python
import requests
header = {'user-key': '61b413ded5143f11855a455e7cf44568'}
response = requests.get('https://developers.zomato.com/api/v2.1/search', headers = header, params = {'lat': 28.297850, 'lon': 77.140670, 'cuisines': [193], 'sort': 'real_distance', 'order': 'asc'})
data = response.json()
data
```

```python
restaurants = data['restaurants']
restaurants
```

```python
for i in range(0, 10):
    print(restaurants[i]['restaurant']['name'], restaurants[i]['restaurant']['user_rating']['aggregate_rating'], restaurants[i]['restaurant']['R']['res_id'], restaurants[i]['restaurant']['location']['locality'])
```
