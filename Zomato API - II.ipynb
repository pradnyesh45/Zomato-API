## The dataset is highly skewed toward the cities included in Delhi-NCR. So, we will summarise all the other cities in Rest of India while those in New Delhi, Ghaziabad, Noida, Gurgaon, Faridabad to Delhi-NCR. Doing this would make our analysis turn toward Delhi-NCR v Rest of India.


### Plot the bar graph of number of restaurants present in Delhi NCR vs Rest of India.

```python
y = [0, 0]
x = ['Delhi NCR', 'Rest of India']
df = df[df['Country Code'] == 1]
city = np.array(df.City)
for i in city:
    if (i == 'New Delhi' or i == 'Ghaziabad' or i == 'Noida' or i == 'Gurgaon' or i == 'Faridabad'):
        y[0] += 1
    else:
        y[1] += 1
plt.bar(x, y)
plt.xlabel('Delhi-NCR vs Rest of India')
plt.ylabel('Number of Restaurants')
plt.show()
```

### Find the cuisines which are not present in restaurant of Delhi NCR but present in rest of India.Check using Zomato API whether this cuisines are actually not served in restaurants of Delhi-NCR or just it due to incomplete dataset.

```python
df = zomato_data.copy()
df = df[df['Country Code'] == 1]
delhi_ncr = set()
rest_of_india = set()
city = np.array(df.City)
cuisine = np.array(df.Cuisines)
for i in range(0, len(city)):
    j = city[i]
    if (j == 'New Delhi' or j == 'Ghaziabad' or j == 'Noida' or j == 'Gurgaon' or j == 'Faridabad'):
        individual_cuisine = cuisine[i].split(', ')
        for element in individual_cuisine:
            delhi_ncr.add(element)
    else:
        individual_cuisine = cuisine[i].split(', ')
        for element in individual_cuisine:
            rest_of_india.add(element)
answer = rest_of_india - delhi_ncr
for i in answer:
    print(i)
```

### Find the weighted restaurant rating of each locality and find out the top 10 localities with more weighted restaurant rating?


#### Weighted Restaurant Rating=Σ (number of votes * rating) / Σ (number of votes) .

```python
df = zomato_data.copy()
df = df[['Locality', 'Aggregate rating', 'Votes']]
weighted_restaurant_rating = []
locality = []
votes = []
ratings = []
for ind in df.index:
    locality.append(df['Locality'][ind])
    ratings.append(df['Aggregate rating'][ind])
    votes.append(df['Votes'][ind])
for i in range(0, len(votes)):
    if (float(votes[i]) == 0):
        weighted_restaurant_rating.append(0)
        continue
    weighted_restaurant_rating.append((float(votes[i]) * float(ratings[i])) / float(votes[i]))
np_locality = np.array(locality)
np_weighted_restaurant_rating = np.array(weighted_restaurant_rating)
np_locality = np_locality[np.argsort(np_weighted_restaurant_rating)]
np_weighted_restaurant_rating = np.sort(np_weighted_restaurant_rating)
for i in range(len(np_locality) - 1, len(np_locality) - 11, -1):
    print(np_locality[i], np_weighted_restaurant_rating[i])
```
