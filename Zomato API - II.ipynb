```python
zomato_data = pd.read_csv('zomato.csv', encoding='latin-1')
zomato_data
```

## The dataset is highly skewed toward the cities included in Delhi-NCR. So, we will summarise all the other cities in Rest of India while those in New Delhi, Ghaziabad, Noida, Gurgaon, Faridabad to Delhi-NCR. Doing this would make our analysis turn toward Delhi-NCR v Rest of India.


### Plot the bar graph of number of restaurants present in Delhi NCR vs Rest of India.

```python
y = [0, 0]
x = ['Delhi NCR', 'Rest of India']
df = df[df['Country Code'] == 1]
city = np.array(df.City)
for i in city:
    if (i == 'New Delhi' or i == 'Ghaziabad' or i == 'Noida' or i == 'Gurgaon' or i == 'Faridabad'):
        y[0] += 1
    else:
        y[1] += 1
plt.bar(x, y)
plt.xlabel('Delhi-NCR vs Rest of India')
plt.ylabel('Number of Restaurants')
plt.show()
```

### Find the cuisines which are not present in restaurant of Delhi NCR but present in rest of India.Check using Zomato API whether this cuisines are actually not served in restaurants of Delhi-NCR or just it due to incomplete dataset.

```python
df = zomato_data.copy()
df = df[df['Country Code'] == 1]
delhi_ncr = set()
rest_of_india = set()
city = np.array(df.City)
cuisine = np.array(df.Cuisines)
for i in range(0, len(city)):
    j = city[i]
    if (j == 'New Delhi' or j == 'Ghaziabad' or j == 'Noida' or j == 'Gurgaon' or j == 'Faridabad'):
        individual_cuisine = cuisine[i].split(', ')
        for element in individual_cuisine:
            delhi_ncr.add(element)
    else:
        individual_cuisine = cuisine[i].split(', ')
        for element in individual_cuisine:
            rest_of_india.add(element)
answer = rest_of_india - delhi_ncr
for i in answer:
    print(i)
```

### Find the top 10 cuisines served by maximum number of restaurants in Delhi NCR and rest of India.

```python
df = zomato_data.copy()
df = df[df['Country Code'] == 1]
delhi_ncr_dictionary = {}
rest_of_india_dictionary = {}
city = np.array(df.City)
cuisine = np.array(df.Cuisines)
for i in range(0, len(city)):
    j = city[i]
    if (j == 'New Delhi' or j == 'Ghaziabad' or j == 'Noida' or j == 'Gurgaon' or j == 'Faridabad'):
        individual_cuisine = cuisine[i].split(', ')
        for element in individual_cuisine:
            delhi_ncr_dictionary[element] = delhi_ncr_dictionary.get(element, 0) + 1
    else:
        individual_cuisine = cuisine[i].split(', ')
        for element in individual_cuisine:
            rest_of_india_dictionary[element] = rest_of_india_dictionary.get(element, 0) + 1
ncr_cuisine = []
ncr_cuisine_number = []
for i in delhi_ncr_dictionary:
    ncr_cuisine.append(i)
    ncr_cuisine_number.append(delhi_ncr_dictionary[i])
np_ncr_cuisine = np.array(ncr_cuisine)
np_ncr_cuisine_number = np.array(ncr_cuisine_number)
np_ncr_cuisine = np_ncr_cuisine[np.argsort(np_ncr_cuisine_number)]
np_ncr_cuisine_number = np.sort(np_ncr_cuisine_number)
rest_cuisine = []
rest_cuisine_number = []
for i in rest_of_india_dictionary:
    rest_cuisine.append(i)
    rest_cuisine_number.append(rest_of_india_dictionary[i])
np_rest_cuisine = np.array(rest_cuisine)
np_rest_cuisine_number = np.array(rest_cuisine_number)
np_rest_cuisine = np_rest_cuisine[np.argsort(np_rest_cuisine_number)]
np_rest_cuisine_number = np.sort(np_rest_cuisine_number)
print('** Delhi NCR **')
number = []
for i in range(len(np_ncr_cuisine) - 1, len(np_ncr_cuisine) - 11, -1):
    number.append(np_ncr_cuisine_number[i])
    print(np_ncr_cuisine[i], np_ncr_cuisine_number[i])
plt.pie(number)
plt.show()
print('** Rest Of India **')
number = []
for i in range(len(np_rest_cuisine) - 1, len(np_rest_cuisine) - 11, -1):
    number.append(np_ncr_cuisine_number[i])
    print(np_rest_cuisine[i], np_rest_cuisine_number[i])
plt.pie(number)
plt.show()
```
