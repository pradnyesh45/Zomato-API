### Fetch the details of "Pa Pa Ya" restaurant using Zomato API.
### Print the user rating, the average cost for two, cuisines and address of "Pa Pa Ya" restaurant.

```python
import requests
header = {'user-key': '61b413ded5143f11855a455e7cf44568'}
response = requests.get('https://developers.zomato.com/api/v2.1/search', headers = header, params = {'q': 'Pa Pa Ya'})
data = response.json()
data
```

```python
restaurants = data['restaurants']
restaurants
```

```python
print(restaurants[0]['restaurant']['user_rating']['aggregate_rating'])
print(restaurants[0]['restaurant']['average_cost_for_two'])
print(restaurants[0]['restaurant']['cuisines'])
print(restaurants[0]['restaurant']['location']['address'])
```

### Find out whether online table reservation is supported in "Pa Pa Ya" restaurant or not using Zomato API.
### Print 'yes' if online table reservation is supported else 'no'
## Note :
### In Zomato response JSON. 1=Yes and 0 = No

```python
import requests
header = {'user-key': '61b413ded5143f11855a455e7cf44568'}
response = requests.get('https://developers.zomato.com/api/v2.1/restaurant', headers = header, params = {'q': 'Pa Pa Ya', 'res_id': 18429148})
data = response.json()
data
```

```python
if (data['has_online_delivery']):
    print('yes')
else:
    print('no')
```

### Fetch the latest review of "Pa Pa Ya" restaurant using Zomato API.
### Print the name of the user, user rating and review text.
## Note :
### Due to the limitation of the basic API plan, it can only fetch 5 latest review

```python
import requests
header = {'user-key': '61b413ded5143f11855a455e7cf44568'}
response = requests.get('https://developers.zomato.com/api/v2.1/reviews', headers = header, params = {'res_id': 18429148})
data = response.json()
data
```

```python
user_reviews = data['user_reviews']
user_reviews
```

```python
for i in range(0, len(user_reviews)):
    print(user_reviews[i]['review']['user']['name'], user_reviews[i]['review']['rating'], user_reviews[i]['review']['review_text'])
```
